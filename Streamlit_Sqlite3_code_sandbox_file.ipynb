```python
import statistics as st
```

```python
ddist = st.NormalDist(1_000,300)
```

```python
mc_dist = ddist.samples(100_000)
```

```python
len(mc_dist)
```

```python
ddist.cdf(1090)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(mc_dist, alpha=0.85)
plt.title("MC simulation")
plt.ylabel("Value")
plt.xlabel("Index")
plt.grid(alpha=0.75, linewidth=0.05, color='black')
plt.show()
```

```python
plt.hist(mc_dist, bins=50, density=True, alpha=0.6, cumulative=False)
plt.grid(alpha=0.45, linewidth=0.15, color='black')
plt.ylabel("Density")
plt.xlabel("Value")
plt.show()
```
