```python
bond_index = pdr.DataReader('WIG20TR.PL', data_source='stooq', start='2010-01-01', end='2025-01-01')
```

```python
bond_index.sort_index(ascending=True, inplace=True)
bond_index_weekly = bond_index.resample('W').last()
bond_index_weekly
```

```python
bond_index_ret = bond_index['Close'].pct_change()
```

```python
bond_index_ret = bond_index_ret.resample('W').sum()
bond_index_ret
```

```python
plt.figure(figsize=(10,3), dpi=300)
plt.plot(bond_index_ret)
plt.title("Bond index returns")
plt.grid(alpha=0.25)
```

```python
plt.figure(figsize=(10,3), dpi=300)
plt.yscale('log')
plt.plot(bond_index['Close'].tail(500))
plt.plot(bond_index['Close'].rolling(250).mean().tail(500))
plt.title("12-month rolling mean")
plt.xlabel("Date")
plt.ylabel("Index value")
plt.grid(alpha=0.25)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(14,6))
ax1.plot(bond_index_ret)
ax1.hlines(y=[bond_index_ret.std(), -bond_index_ret.std()],
           xmin=bond_index_ret.index.min(),
           xmax=bond_index_ret.index.max(),
           linestyles='solid',
           color='red')
ax1.hlines(y=[bond_index_ret.std()*2, -bond_index_ret.std()*2],
           xmin=bond_index_ret.index.min(),
           xmax=bond_index_ret.index.max(),
           linestyles='solid',
           color='yellow',
           alpha=0.95, lw=0.5)
ax1.grid(alpha=0.25)
ax1.set_title("Bond index returns")
ax2.set_title("Bond index")
ax2.plot(bond_index['Close'].rolling(52).mean())
ax2.grid(alpha=0.25)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create the first plot
plt.figure(figsize=(8, 4))
plt.plot([1, 2, 3], [3, 2, 1], label='Line Graph')
plt.title('Plot 1')
plt.xlabel('X axis 1')
plt.ylabel('Y axis 1')
plt.legend()
plt.grid(alpha=0.5)

# Create the second plot
plt.figure(figsize=(8, 4))
plt.scatter([1, 2, 3], [3, 2, 1], color='red', label='Scatter Plot')
plt.title('Plot 2')
plt.xlabel('X axis 2')
plt.ylabel('Y axis 2')
plt.legend()
plt.grid(alpha=0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

# First line with ax1
ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='tab:blue')
ax1.plot([1, 2, 3], [3, 2, 1], color='tab:blue', label='Line 1')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.set_ylabel('Y2 data', color='tab:red')
ax2.plot([1, 2, 3], [1, 4, 9], color='tab:red', linestyle='--', label='Line 2')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Two y-axis plot')
fig.tight_layout()
plt.show()
```
